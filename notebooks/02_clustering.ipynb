# Klustring — UMAP + KMeans
Visualisera fotostilar i CLIP-embedding-rymden.

```python
import numpy as np
import pandas as pd
import plotly.express as px

df = pd.read_csv('../data/processed/clustered.csv')
print(df.shape, df.columns.tolist())
```

```python
fig = px.scatter(
    df, x='umap_x', y='umap_y',
    color='cluster', hover_data=['title', 'score'],
    title='UMAP 2D — Stilkluster',
    color_continuous_scale='rainbow',
)
fig.show()
```

```python
# Score per cluster — vilka stilar premieras?
cluster_stats = df.groupby('cluster')['score'].agg(['mean', 'median', 'count'])
cluster_stats.sort_values('mean', ascending=False)
```

```python
import matplotlib.pyplot as plt
from PIL import Image

# Visa top-3 bilder per kluster
n_clusters = df['cluster'].nunique()
fig, axes = plt.subplots(n_clusters, 3, figsize=(12, n_clusters * 4))

for cluster_id in range(n_clusters):
    top = df[df['cluster'] == cluster_id].nlargest(3, 'score')
    for col, (_, row) in enumerate(top.iterrows()):
        ax = axes[cluster_id][col]
        try:
            img = Image.open(row['local_path'])
            ax.imshow(img)
            ax.set_title(f"score={row['score']}", fontsize=8)
        except Exception:
            ax.axis('off')
        ax.axis('off')
    axes[cluster_id][0].set_ylabel(f'Cluster {cluster_id}', rotation=0, labelpad=60)

plt.suptitle('Top photos per cluster', fontsize=14)
plt.tight_layout()
```
